```python
import yfinance as yf
import pandas as pd
```

```python
tickers = ["CPRX", "UTHR", "INCY", "QGEN"]
```

```python
data = {}
for i in tickers: 
    ticker= yf.Ticker(i)
    df = ticker.history('5Y')
    data[i] = df
```

```python
data = pd.concat(data, axis = 0, ignore_index = True)
```
