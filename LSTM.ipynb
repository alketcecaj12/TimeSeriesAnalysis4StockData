```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
```

```python
# Generate a synthetic sine wave dataset
def generate_sine_wave_data(seq_length, num_samples):
    x = np.linspace(0, 100, num_samples)
    y = np.sin(x)
    return y

# Parameters
seq_length = 50  # Length of the sequence used for prediction
num_samples = 1000  # Number of total samples

# Generate the data
data = generate_sine_wave_data(seq_length, num_samples)

# Scaling the data
scaler = MinMaxScaler(feature_range=(0, 1))
data = scaler.fit_transform(data.reshape(-1, 1))

# Create sequences of input-output pairs
def create_sequences(data, seq_length):
    X = []
    y = []
    for i in range(len(data) - seq_length):
        X.append(data[i:i + seq_length])
        y.append(data[i + seq_length])
    return np.array(X), np.array(y)

# Prepare the data
X, y = create_sequences(data, seq_length)

# Reshape the input to be [samples, time steps, features]
X = X.reshape((X.shape[0], X.shape[1], 1))

# Define the LSTM model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(seq_length, 1)))
model.add(LSTM(50, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X, y, batch_size=1, epochs=10)

# Make predictions
predicted = model.predict(X)
```

```python
# Inverse transform the predictions
predicted = scaler.inverse_transform(predicted)
actual = scaler.inverse_transform(y.reshape(-1, 1))
```

```python
# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(actual, label='Actual Data')
plt.plot(predicted, label='Predicted Data')
plt.title('Time Series Prediction')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()
```
